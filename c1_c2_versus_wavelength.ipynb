```python
from astropy.io import ascii
import numpy as np, matplotlib.pyplot as plt


t = ascii.read('quadratic-ld-5777+4.43+0.0-209wavelengths.ecsv')
w = ascii.read('quadratic-ld-5400+4.45+0.0-209wavelengths.ecsv')
ws = ascii.read('wasp39b-nirspec-prism-coefficients.ecsv')
```

```python
plt.figure(figsize=(6,4), dpi=200)
plt.plot(t['wavelength'], t['c1'], label='$c_1$ theoretical Sun', color='red')
plt.plot(t['wavelength'], t['c2'], label='$c_2$ theoretical Sun', color='orange')
plt.plot(w['wavelength'], w['c1'], label='$c_1$ theoretical WASP39', color='green')
plt.plot(w['wavelength'], w['c2'], label='$c_2$ theoretical WASP39', color='lightgreen')

plt.ylim(0,0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Quadratic Limb-Darkening')
plt.title('Theoretical Limb-Darkening Coeffecients Versus Wavelength')
plt.legend(frameon=False);
```

```python
tt = t['c1']+t['c2']
wt = w['c1']+w['c2']
```

```python

plt.figure(figsize=(6,4), dpi=200)
plt.plot(t['wavelength'], tt, label='$c_1+c_2$ theoretical Sun', color='red')
plt.plot(w['wavelength'], wt, label='$c_1+c_2$ theoretical WASP39', color='green')

plt.ylim(0,0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Quadratic Limb-Darkening')
plt.title('Theoretical Limb-Darkening Coeffecients Versus Wavelength')
plt.legend(frameon=False);

```

```python
plt.figure(figsize=(6,4), dpi=200)
plt.plot(ws['wavelength'], ws['c2'], label='$c_2$ empirical WASP39', color='lightblue')
plt.plot(ws['wavelength'], ws['c1'], label='$c_1$ empirical WASP39', color='blue')


plt.ylim(0,0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Quadratic Limb-Darkening')
plt.title('Empirical Limb-Darkening Coeffecients Versus Wavelength')
plt.legend(frameon=False);
```

```python
z = np.polyfit(ws['wavelength'], ws['c1+c2'], 2)
p = np.poly1d(z)

plt.figure(figsize=(6,4), dpi=200)
plt.plot(ws['wavelength'], ws['c1+c2'], label='$c_1+c_2$ WASP-39')
plt.plot(ws['wavelength'], p(ws['wavelength']),label='$(2.307*10^{-8})\lambda^2 + (-2.172*10^{-4})\lambda + (6.862*10^{-1})$')

print(z)
plt.ylim(0,0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Quadratic Limb-Darkening')
plt.title('Empirical Limb-Darkening Coeffecients Versus Wavelength')
plt.legend(frameon=False);
```

```python
plt.figure(figsize=(6,4), dpi=200)
plt.plot(ws['wavelength'], ws['c1+c2'], label='c1+c2 empricial WASP-39')
plt.plot(t['wavelength'], tt, label='$c_1+c_2$ theoretical Sun', color='red')
plt.plot(w['wavelength'], wt, label='$c_1+c_2$ theoretical WASP39', color='green')


plt.ylim(0,0.7)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Quadratic Limb-Darkening')
plt.title('Limb-Darkening Coeffecients($c_1+c_2$) Versus Wavelength')
plt.legend(frameon=False);

```
