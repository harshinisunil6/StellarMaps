```python
import LimbsStellarMap as lsm
import matplotlib.pyplot as plt
```

This function takes two limb-darkening coefficients ($c_1$ and $c_2$) to make an intensity map of star using a quadratic limb-darkening model. The equation for the intensity is 
$$I(\mu)=I_{0}[1 - c_1 (1- \mu) -c_2(1-\mu)^{2}]$$

```python
lsm.intensity_imshow(.4,.2, 'intensity')
plt.savefig('example.png')
```
